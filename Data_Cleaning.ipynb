```python
import pandas as pd
data=pd.read_csv("/content/sample_data/JSAMP_pisem_SS.csv")
```

```python
ports=[80,443,8080,8443]
data=data[data.port.isin(ports)]
data=data[data.domain.str.contains('jio.com')]
print(data)
```

```python
data=data[~data.domain.str.contains('api')]
print(len(data))
```

```python
print(data.public_ip.nunique())
```

```python
idx = data['public_ip'].drop_duplicates().index
#this will return filtered df
data=data.loc[idx,:]
```

```python
print(len(data))
```

```python
print(data)
```

```python
data.to_csv('data.csv', header=True, index=False)
```

```python
import pandas as pd
dd=pd.read_csv("/content/data.csv")
```

```python
dd['new']=dd['public_ip'].astype(str)+':'+dd['port'].astype(str)
```

```python
print(dd)
```

```python
print(str(dd.loc[1,['new']]))
```

```python



dd.loc[(dd['port'] != 80) & (dd['port'] != 443), 'final'] = 'https://'+dd['public_ip']+':'+dd['port'].astype(str)
dd.loc[dd['port'] == 80, 'final'] = 'http://'+dd['public_ip']
dd.loc[dd['port'] == 443, 'final'] = 'https://'+dd['public_ip']
```

```python
print(len(dd))
```

```python
for i in dd['final']:
  print(i)
```

```python
for i in dd['port']:
  print(i)
```
